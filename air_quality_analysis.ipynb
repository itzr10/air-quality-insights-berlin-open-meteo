```python
# ================================================================
# 0. Setup: Install & Imports
# ================================================================

import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta, timezone

plt.style.use("seaborn-v0_8")
pd.set_option("display.max_columns", None)

print("✓ Imports successful")

# ================================================================
# 1. Configuration
# ================================================================

# Berlin coordinates
LAT = 52.52
LON = 13.405

# Time range: last 7 days up to today
end_date = datetime.now(timezone.utc).date()
start_date = end_date - timedelta(days=7)

start_str = start_date.isoformat()
end_str = end_date.isoformat()

print(f"Fetching data for Berlin from {start_str} to {end_str}")

# Open-Meteo API endpoints (no API key required)
AIR_QUALITY_URL = "https://air-quality-api.open-meteo.com/v1/air-quality"
WEATHER_URL = "https://api.open-meteo.com/v1/forecast"

# ================================================================
# 2. Fetch Air Quality Data
# Docs: https://open-meteo.com/en/docs/air-quality-api
# ================================================================

air_params = {
    "latitude": LAT,
    "longitude": LON,
    "start_date": start_str,
    "end_date": end_str,
    "hourly": [
        "pm2_5",
        "pm10",
        "ozone",
        "nitrogen_dioxide",
        "sulphur_dioxide",
        "carbon_monoxide"
    ],
    "timezone": "auto",
}

resp_air = requests.get(AIR_QUALITY_URL, params=air_params, timeout=30)
resp_air.raise_for_status()
air_json = resp_air.json()

# Convert to DataFrame
air_hourly = air_json["hourly"]
df_air = pd.DataFrame(air_hourly)
df_air["time"] = pd.to_datetime(df_air["time"])
df_air = df_air.set_index("time").sort_index()

print("✓ Air quality data shape:", df_air.shape)
display(df_air.head())

# ================================================================
# 3. Fetch Weather Data
# Docs: https://open-meteo.com/en/docs
# ================================================================

weather_params = {
    "latitude": LAT,
    "longitude": LON,
    "start_date": start_str,
    "end_date": end_str,
    "hourly": [
        "temperature_2m",
        "relative_humidity_2m",
        "wind_speed_10m",
        "precipitation"
    ],
    "timezone": "auto",
}

resp_weather = requests.get(WEATHER_URL, params=weather_params, timeout=30)
resp_weather.raise_for_status()
weather_json = resp_weather.json()

weather_hourly = weather_json["hourly"]
df_weather = pd.DataFrame(weather_hourly)
df_weather["time"] = pd.to_datetime(df_weather["time"])
df_weather = df_weather.set_index("time").sort_index()

print("✓ Weather data shape:", df_weather.shape)
display(df_weather.head())

# ================================================================
# 4. Merge Air Quality & Weather
# ================================================================

df = df_air.join(df_weather, how="inner")
df = df.sort_index()

print("✓ Merged data shape:", df.shape)
display(df.head())

# Basic cleaning: ensure numeric types
df = df.apply(pd.to_numeric, errors="coerce")

# Drop rows that are completely empty
df = df.dropna(how="all")

print("After cleaning:", df.shape)

# ================================================================
# 5. Feature Engineering
# - Approximate AQI bucket from PM2.5 (very simplified)
# - Discomfort index from temperature & humidity
# ================================================================

def pm25_bucket(pm):
    if pd.isna(pm):
        return "Unknown"
    if pm <= 12:
        return "Good"
    elif pm <= 35.4:
        return "Moderate"
    elif pm <= 55.4:
        return "Unhealthy for Sensitive"
    elif pm <= 150.4:
        return "Unhealthy"
    elif pm <= 250.4:
        return "Very Unhealthy"
    else:
        return "Hazardous"

df["pm25_bucket"] = df["pm2_5"].apply(pm25_bucket)

# Simple discomfort index (not clinical, just illustrative)
# DI ~ T - 0.55*(1 - RH)* (T - 14.5), RH in fraction
rh_frac = df["relative_humidity_2m"] / 100.0
df["discomfort_index"] = (
    df["temperature_2m"] - 0.55 * (1 - rh_frac) * (df["temperature_2m"] - 14.5)
)

display(df[["pm2_5", "pm25_bucket", "temperature_2m", "relative_humidity_2m", "discomfort_index"]].head())

# ================================================================
# 6. Basic Exploratory Data Analysis
# ================================================================

print("\n=== Summary Statistics ===")
display(df.describe().T)

print("\n=== PM2.5 Bucket Counts ===")
print(df["pm25_bucket"].value_counts(dropna=False))

# Correlation between pollution and weather
cols_of_interest = [
    "pm2_5",
    "pm10",
    "ozone",
    "nitrogen_dioxide",
    "temperature_2m",
    "relative_humidity_2m",
    "wind_speed_10m",
    "precipitation",
    "discomfort_index"
]
corr = df[cols_of_interest].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation between Air Quality & Weather Variables (Berlin)")
plt.tight_layout()
plt.show()

# ================================================================
# 7. Visualization Examples
# ================================================================

# Time series of PM2.5
plt.figure(figsize=(12, 4))
df["pm2_5"].plot()
plt.title("PM2.5 Concentration Over Time (Berlin)")
plt.ylabel("PM2.5 (µg/m³)")
plt.xlabel("Time")
plt.tight_layout()
plt.show()

# Boxplot of PM2.5 by hour of day
df["hour"] = df.index.hour
plt.figure(figsize=(10, 4))
sns.boxplot(x="hour", y="pm2_5", data=df)
plt.title("PM2.5 by Hour of Day (Berlin, last 7 days)")
plt.ylabel("PM2.5 (µg/m³)")
plt.xlabel("Hour of Day")
plt.tight_layout()
plt.show()

# Scatter plot: PM2.5 vs Wind Speed
plt.figure(figsize=(6, 4))
sns.scatterplot(x="wind_speed_10m", y="pm2_5", data=df, alpha=0.5)
plt.title("PM2.5 vs Wind Speed (Berlin)")
plt.xlabel("Wind Speed (m/s)")
plt.ylabel("PM2.5 (µg/m³)")
plt.tight_layout()
plt.show()

# ================================================================
# 8. Save Outputs
# ================================================================

output_csv = "berlin_air_quality_weather_last7days.csv"
df.to_csv(output_csv, index=True)
print(f"✓ Saved merged dataset to {output_csv}")

# Optionally save a correlation heatmap as PNG
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation between Air Quality & Weather Variables (Berlin)")
plt.tight_layout()
heatmap_file = "correlation_heatmap_berlin.png"
plt.savefig(heatmap_file, dpi=150)
plt.close()
print(f"✓ Saved heatmap to {heatmap_file}")

print("\nPipeline completed successfully ✅")
```
